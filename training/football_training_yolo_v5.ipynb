```python
from roboflow import Roboflow
rf = Roboflow(api_key="nlwl5GCHDngWYc9pJgHG")
project = rf.workspace("roboflow-jvuqo").project("football-players-detection-3zvbc")
version = project.version(1)
dataset = version.download("yolov5")
```

```python
import os
import shutil

# source path from roboflow dataset object
src = dataset.location  # existing dataset folder

# destination folder ./training/<dataset_folder_name>
dst_root = os.path.join(".", "training")
os.makedirs(dst_root, exist_ok=True)
dst = os.path.join(dst_root, os.path.basename(os.path.normpath(src)))

# if destination exists, remove it to ensure a clean copy
if os.path.abspath(src) == os.path.abspath(dst):
    print(f"Dataset already at {dst}")
else:
    if os.path.exists(dst):
        shutil.rmtree(dst)
    shutil.copytree(src, dst)

    # try to update the dataset.location so subsequent cells use the new path
    try:
        dataset.location = dst
    except Exception:
        pass

    print(f"Dataset moved/copied to {dst}")
```
